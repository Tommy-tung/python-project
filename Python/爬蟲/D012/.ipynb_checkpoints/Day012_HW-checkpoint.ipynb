### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_= 'part_list_2').find_all('h3') : 
    print(d.find(class_= 'date').text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
cat = []
for d in soup.find(class_= 'part_list_2').find_all('h3') : 
    date = d.find(class_= 'date').text
    tag = d.find_all('em')[0].text
    target = d.find_all('a')[-1].text
    cat.append([date,tag,target])
cat = pd.DataFrame(cat, columns = ['date', 'tag', 'title'])
cat
```

### ③ 哪一個類別的文章最多

```python
df = pd.DataFrame(cat.groupby('tag')['tag'].count())
df 
```
