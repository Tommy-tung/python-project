## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.ptt.cc/bbs/NBA/M.1576394531.A.C02.html'
resp = requests.get(url)
print(resp.text)
```

```python
soup = BeautifulSoup(resp.text)
print(soup.title)
```

## Grab + PyQuery

```python
from grab import Grab
from pyquery import PyQuery as pq
g = Grab()
resp = g.go('https://www.ptt.cc/bbs/NBA/M.1576394531.A.C02.html')
resp.body
```

```python
doc = pq(resp.body.decode('utf-8'))
print(doc)
```
