```python
df = statis_df.pivot(index = 'ID', columns = '題目', values = '答案')
df = df.reset_index()
#df.columns = df.columns.tolist()
#df = df.merge(info_df, on = 'ID')
df = df.merge(personal_df, on = 'ID')
```

```python
columns_name = ['ID','ques1', 'ques10', 'ques11', 'ques12', 'ques13', 'ques14', 
                'ques2', 'ques3', 'ques4', 'ques5', 'ques6', 'ques7', 'ques8','ques9',
                '性別','年齡','教育程度','職業類別'  ]
df.columns = columns_name
```

```python
name = ['ID']
ques_name = []
info_name = ['性別','年齡','教育程度','職業類別']
for i in range(14) : 
    ques_name.append('ques' + str(i+1))
#ques_name.extend(['雙連安養中心','荷蘭失智村'])
name.extend(ques_name)
name.extend(info_name)
name
df = df[name]
```

```python
df
```

```python
a = df.iloc[:, [1,-1]]
```

```python
a.columns = ['ques1', '職業類別']
```

```python
mod = ols('ques1 ~ 職業類別', data = a).fit()
ans = sm.stats.anova_lm(mod, typ = 2).iloc[0,2:].tolist()
```

```python
ans
```

```python
sm.stats.anova_lm(mod, typ = 2).iloc[0,2:]
```

```python
result = []
for i in ques_name : 
    for j in info_name : 
        val = []
        val.append(i)
        val.append(j)
        df_anl = df[val]
        mod = ols(i + ' ~ ' + j, data = df_anl).fit()
        ans = sm.stats.anova_lm(mod, typ = 2).iloc[0,2:].tolist()
        ans.append(i)
        ans.append(j)
        result.append(ans)
```

```python
result = pd.DataFrame(result)
result.columns = ['F值', 'p-value', '問題','個人資料']
result = result[['問題', '個人資料', 'F值', 'p-value']]
```
