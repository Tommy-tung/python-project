```python
import requests 
import json 
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
json.loads(response)
```

```python
print(json.loads(response)[0]['title'])
json.loads(response)[1]['createdAt']
```
